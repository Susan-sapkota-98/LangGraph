```python
from langgraph.graph import StateGraph,START,END
from langchain_ollama import ChatOllama
from typing import TypedDict
```

```python
model = ChatOllama(model="llama3.2")  
```

```python
class BlogState(TypedDict):

    title: str
    outline: str
    content: str
    evaluate: str
    
    
```

```python
def create_outline(state: BlogState) -> BlogState:

    # fetch title
    title=state['title']

    # call llm gen outline
    prompt=f'Generate a detailed outline for a blog on the topic - {title}'
    outline=model.invoke(prompt).content

    #update state
    state['outline']=outline

    return state
```

```python
def create_blog(state: BlogState) -> BlogState:

    # fetch title
    title=state['title']


    # call llm gen outline
    outline=state['outline']
    prompt=f'Write a detailed blog on the title - {title} using the following outline \n {outline}'
    content=model.invoke(prompt).content

    #update state
    state['content']=content

    return state
```

```python
def create_evaluate(state: BlogState) -> BlogState:

    # fetch title
    title=state['title']


    # call llm gen outline
    outline=state['outline']
    content=state['content']
    prompt=f'based on Outline - {outline} rate my blog \n {content}'
    evaluate=model.invoke(prompt).content

    #update state
    state['evaluate']=evaluate

    return state
```

```python
graph=StateGraph(BlogState)

# nodes
graph.add_node('create_outline',create_outline)
graph.add_node('create_blog',create_blog)
graph.add_node('create_evaluate',create_evaluate)

# edges

graph.add_edge(START,'create_outline')
graph.add_edge('create_outline','create_blog')
graph.add_edge('create_blog','create_evaluate')
graph.add_edge('create_evaluate',END)

workflow=graph.compile()
```

```python
# execute

final_state = workflow.invoke({'title': 'Rise of AI in Nepal'})
print(final_state['content']) # only for answer
```
